```python
import os
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
df = pd.read_csv("/content/drive/MyDrive/Mobile Shopee/cleaned_data1.csv")
df.head()
```

```python
cols = ["Brand", "operating_system", "Processor"]
transactions = df[cols]
transactions.head()
```

```python
df_encoded = pd.get_dummies(transactions)
df_encoded.head()
```

```python
frequent_items = fpgrowth(df_encoded, min_support=0.01, use_colnames=True)
frequent_items.head()
```

```python
rules = association_rules(frequent_items,
                          metric="confidence",
                          min_threshold=0.3)

rules = rules.sort_values("confidence", ascending=False)
rules.head()
```

```python
def recommend(user_items, rules):

    user_items = set(user_items)
    results = []

    for _, row in rules.iterrows():
        antecedent = set(row['antecedents'])
        consequent = set(row['consequents'])

        if antecedent.issubset(user_items):
            results.append([
                list(consequent)[0],
                row['confidence'],
                row['lift']
            ])

    return results
```

```python
user_input = ["Brand_Samsung", "operating_system_Android"]
recs = recommend(user_input, rules)

print("Recommended Items:\n")
for rec in recs:
    print(f"Recommended: {rec[0]} | Confidence: {rec[1]:.3f} | Lift: {rec[2]:.3f}")
```
